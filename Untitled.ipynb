```python
import pandas as pd
Flyline = pd.read_csv('Flyline_FBg_whole_genome.csv')
Flyline = Flyline.drop(['Collection', 'FlyBase link', 'Gene'], axis = 1)
Mousefile = pd.read_csv('/Users/joshtitlow/Downloads/mart_export2.txt', sep = '\t')
Mousename = pd.read_csv('/Users/joshtitlow/Downloads/mart_export.txt', sep = '\t')

Mousefile = Mousefile.drop(['Transcript stable ID'], axis = 1)
Mousefile = Mousefile.dropna()
merge = pd.merge(Mousefile, Mousename, how='left', on=['Gene stable ID'])
merge = merge.dropna()

merge = merge.groupby('Drosophila melanogaster gene stable ID')['Gene name'].apply(' '.join).reset_index()

Unique = []
for index, row in merge.iterrows():
    #print row['Drosophila melanogaster gene stable ID']
    genes = []
    genenames = str(list(set(row['Gene name'].split()))).strip('[]')
    genes.append(genenames)
    Unique.append(str(genes)[2:-2].replace('''''', ))

# add lists to ref_file
merge['Unique'] = Unique 
merge.to_csv('merge.csv', index=False)

```
